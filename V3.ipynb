```python
# Zodiac sign
# Stany Vanhemelrijck
# VAN20500599
# Last updated on November 17th 2021

# This line of numbers is to count up to 79 characters.
# Following the guidelines of pep 8, the maximum number of characters on 
# one line increases readability on smaller screens or when screens are 
# reduced in size. 
# The only place where this was not followed was when a long string was 
# printed out. In that case, clear formatting of what the user sees was chosen
# above the readability of the code.
#234567891123456789212345678931234567894123456789512345678961234567897123456789

# Date from the datetime module is used to make calculations with dates. 
from datetime import date
# The pandas module is used to create a table (called a DataFrame in pandas)
import pandas


def calculate_age(input_dob):
    """This function calculates the age using the date of birth inputted 
    by the user.
    """
    # Converting the input to a date using the datetime module. 
    # .fromisoformat() transforms a string into a date type.
    dob = date.fromisoformat(input_dob)
        
    # .today() from the datetime module is used to determine today's date.
    today = date.today()

    # Calculating the age by subtracting the year of the date of birth 
    # by the current year.
    age = today.year - dob.year 

    # Defining the real age by considering the day and month: 
    # if the date today is less than the bith month and day, 
    # meaning the birthday of this year is still to come, subtract 1 year. 
    # in other words in the year 2021 I should be 26 but my birthday is on the
    # 26th of November, so until that day I'm 25.

    # ****
    # there is a shorter way for using this logic: 
    # by subtracting the statement: 
    # ((today.month, today.day) < (dob.month, dob.day))
    # from 
    # today.year - dob.year.
    # if today.month and today.day is smaller than dob.month, dob.day,
    # python returns True or 1 and 
    # it will subtract 1 from today.year - dob.year 
    # For purpose of having clear code I choose to stick with the if statement. 
    # ****
    
    # the day and month today is smaller than the day and month of the birthday 
    # meaning the birthday of the user is still to come, we subtract 
    # 1 year of the age. Otherwise we keep the original age. 
    if ((today.month, today.day) < (dob.month, dob.day)):
        final_age = age -1
    else: 
        final_age = age

    return final_age


def birthday(input_dob):
    """Replace the year in the dob with the current year and 
    adding it to the variable birthday."""
    # Putting the the dob from the user in the right format. 
    dob = date.fromisoformat(input_dob)
    # Defining the date of today.
    today = date.today()
    birthday = dob.replace(year = today.year)
    
    return birthday


def zodiac_sign(birthday_of_person):
    """A function that determines the zodiac sign using the birthday."""
    # Defining the date of today
    today = date.today()
    # if the birthday is between one date and another date, that zodiac sign is defined. 
    if birthday_of_person >= date(today.year, 3, 21) and birthday_of_person <= date(today.year, 4, 19):
        zodiac_sign = "Aries"
    elif birthday_of_person >= date(today.year, 4, 20) and birthday_of_person <= date(today.year, 4, 20):
        zodiac_sign = "Taurus"
    elif birthday_of_person >= date(today.year, 5, 21) and birthday_of_person <= date(today.year, 6, 20):
        zodiac_sign = "Gemini"
    elif birthday_of_person >= date(today.year, 6, 21) and birthday_of_person <= date(today.year, 7, 22):
        zodiac_sign = "Cancer"
    elif birthday_of_person >= date(today.year, 7, 23) and birthday_of_person <= date(today.year, 8, 22):
        zodiac_sign = "Leo"
    elif birthday_of_person >= date(today.year, 8, 23) and birthday_of_person <= date(today.year, 9, 22):
        zodiac_sign = "Virgo"
    elif birthday_of_person >= date(today.year, 9, 23) and birthday_of_person <= date(today.year, 10, 22):
        zodiac_sign = "Libra"
    elif birthday_of_person >= date(today.year, 9, 23) and birthday_of_person <= date(today.year, 11, 21):
        zodiac_sign = "Scorpio"
    elif birthday_of_person >= date(today.year, 11, 22) and birthday_of_person <= date(today.year, 12, 21):
        zodiac_sign = "Sagittarius"
    elif birthday_of_person >= date(today.year, 12, 22) and birthday_of_person <= date(today.year, 1, 19):
        zodiac_sign = "Capricorn"
    elif birthday_of_person >= date(today.year, 1, 20) and birthday_of_person <= date(today.year, 2, 18):
        zodiac_sign = "Aquarius"
    elif birthday_of_person >= date(today.year, 2, 19) and birthday_of_person <= date(today.year, 3, 20):
        zodiac_sign = "Pisces"
    else:
        print("Zodiac sign not found")
        
    return zodiac_sign

#234567891123456789212345678931234567894123456789512345678961234567897123456789
# MAIN PROGRAM
# Creating an empty list that will hold all entries of the user. 
people = []

def main():
    """ This is the main program, that ask the user for input. 
    It will keep on looping until the user inputs 'end', 
    in the id_num variable."""
    # Defining end to be zero (or off).
    # conditional is made.
    end = 0
    # While end is not equal to 1 the loop can run.
    while end != 1: 
        print("Whenever you like to stop, enter 'end' into User ID.")
        # Asking input for the id number from the user.
        id_num = input("Input your user ID, and press enter: ")
        
        # if the input for user ID is 'end' the loop stops. 
        if id_num == 'end':
            # set the conditional to 1 
            end = 1
        # id the input is not end:
        else:
            # Ask for a name input
            name = input("Input your name, and press enter:  ")
            # Ask for a date of birth.
            input_dob = input("Input your date of birth (YYYY-MM-DD) including 0's, and press enter: ")
        
            # Saving the first input set to a list 'person'.
            person = [id_num, name, input_dob]
            
            # Runs the calculate_age function to calculate the age. 
            age = str(calculate_age(input_dob))
            # Adds the age to the person list. 
            person.append(age)
            
            #Calculates the birthday, that is used in the zodiac_sign function. 
            birthday_of_person = birthday(input_dob)
            
            zodiac_sign_of_person = zodiac_sign(birthday_of_person)
            person.append(zodiac_sign_of_person)
            
            # Add person to people.
            people.append(person)
    return people

# updating the people list by running the main function.
people = main()
print(people)

# Printing a table using the pandas library.
pandas.DataFrame(people, columns = ['UID', 'Name', 'D.o.B.', 'Age', 'Western Zodiac'])
```
